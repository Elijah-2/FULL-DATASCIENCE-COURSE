# 11. Subplots
#### To create multiple plots in one figure.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

axes[0].plot(x, y, color="red")
axes[0].set_title("Line Plot")

axes[1].bar(categories, values, color="green")
axes[1].set_title("Bar Chart")

axes[2].bar(categories, values, color="green")
axes[2].set_title("Bar Chart")

plt.show()
```

```python
fig: Represents the entire figure (a container for all subplots).
axes: Represents the individual subplot objects inside the figure.
plt.subplots(1, 2): Creates 1 row and 2 columns of subplots (i.e., two side-by-side plots).
figsize=(10, 5): Sets the figure size to 10 inches wide and 5 inches tall.
```

# 13. Adding Annotations
### To highlight points on the graph.

```python
plt.plot(x, y)
plt.annotate("Important Point", xy=(3, 25), xytext=(4, 30), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
Understanding plt.annotate() in Matplotlib
The plt.annotate() function is used to add text annotations to specific points in a plot. It helps highlight important points by adding labels, arrows, or description
```

```python
plt.annotate(
    "Important Point",  # Text label
    xy=(3, 25),         # The point to annotate (x=3, y=25)
    xytext=(4, 30),     # Position of the annotation text (x=4, y=30)
    arrowprops=dict(facecolor='black', shrink=0.05)  # Arrow properties
)
```

```python
 xy=(3, 25)  The actual data point being highlighted.
 xytext=(4, 30)  The location where the text label will appear.
 arrowprops  Defines the arrows appearance:

facecolor='black'  The arrow is black.
shrink=0.05  Shrinks the arrow slightly to avoid overlapping.
```

# 14. Filling Area Between Curves
#### Useful for highlighting a region.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color='gray', alpha=0.5)
plt.plot(x, y1, label="sin(x)", color="blue")
plt.plot(x, y2, label="cos(x)", color="red")

plt.legend()
plt.show()
```

# 15. 3D Plotting
#### Using mpl_toolkits.mplot3d

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

```python
Understanding 3D Surface Plots in Matplotlib
The provided code generates a 3D surface plot using mpl_toolkits.mplot3d. This type of plot is useful for visualizing functions of two variables, such as Z = f(X, Y).
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Import 3D plotting toolkit
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Define x and y values (ranges from -5 to 5)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

# Create a meshgrid (2D grid of x, y values)
X, Y = np.meshgrid(x, y)

# Define the Z values using a mathematical function
Z = np.sin(np.sqrt(X**2 + Y**2))  # Generates a ripple effect

# Create a surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

# Show the figure
plt.show()
```

```python
How This Works:
fig = plt.figure()  Creates a new figure.
ax = fig.add_subplot(111, projection='3d')  Adds a 3D subplot.
np.linspace(-5, 5, 50)  Creates 50 evenly spaced values between -5 and 5 for both x and y.
np.meshgrid(x, y)  Creates a grid of X and Y values.
Z = np.sin(np.sqrt(X**2 + Y**2))  Defines Z values as a function of X and Y.
ax.plot_surface(X, Y, Z, cmap='viridis')  Creates a 3D surface plot with a Viridis color map.
plt.show()  Displays the plot.
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Import 3D plotting toolkit
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Define x and y values (ranges from -5 to 5)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

# Create a meshgrid (2D grid of x, y values)
X, Y = np.meshgrid(x, y)

# Define the Z values using a mathematical function
Z = np.tan(np.sqrt(X**3 + Y**3))  # Generates a ripple effect

# Create a surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

# Show the figure
plt.show()
```

# 19. Twin Axes
#### To plot two different scales on the same figure

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y, 'g-')
ax2.plot(x, [i*2 for i in y], 'b-')

ax1.set_xlabel("X-axis")
ax1.set_ylabel("Left Y-axis", color='g')
ax2.set_ylabel("Right Y-axis", color='b')

plt.show()
```

### Understanding twinx() in Matplotlib
### The function ax1.twinx() creates a secondary Y-axis on the right side of the plot. This is useful when plotting two different datasets that share the same X-axis but have different scales.

```python
fig, ax1 = plt.subplots()  # Create a figure and a primary Y-axis

ax2 = ax1.twinx()  # Create a secondary Y-axis (on the right)

ax1.plot(x, y, 'g-')  # Plot on the primary Y-axis (green)
ax2.plot(x, [i*2 for i in y], 'b-')  # Plot on the secondary Y-axis (blue)

ax1.set_xlabel("X-axis")  # Label for the X-axis
ax1.set_ylabel("Left Y-axis", color='g')  # Label for the primary Y-axis (green)
ax2.set_ylabel("Right Y-axis", color='b')  # Label for the secondary Y-axis (blue)

plt.show()  # Display the plot
```

```python
fig, ax1 = plt.subplots()  Creates the figure and the first (primary) Y-axis.
ax2 = ax1.twinx()  Creates a second Y-axis that shares the same X-axis.
ax1.plot(x, y, 'g-')  Plots data on the first Y-axis in green.
ax2.plot(x, [i*2 for i in y], 'b-')  Plots data on the second Y-axis in blue.
ax1.set_ylabel("Left Y-axis", color='g')  Sets a green label for the left Y-axis.
ax2.set_ylabel("Right Y-axis", color='b')  Sets a blue label for the right Y-axis.
plt.show()  Displays the figure.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # First dataset (primary Y-axis)
y2 = np.cos(x) * 2  # Second dataset (secondary Y-axis)

# Create figure and primary Y-axis
fig, ax1 = plt.subplots()

# Create secondary Y-axis
ax2 = ax1.twinx()

# Plot first dataset
ax1.plot(x, y1, 'g-', label="Sine Wave")
ax1.set_ylabel("Sine (Left Y-axis)", color='g')

# Plot second dataset
ax2.plot(x, y2, 'b-', label="Cosine Wave (scaled)")
ax2.set_ylabel("Cosine (Right Y-axis)", color='b')

# Set common X-axis label
ax1.set_xlabel("X-axis")

# Add grid
ax1.grid(True, linestyle="--", alpha=0.6)

# Show legend
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Display the plot
plt.show()
```

# 20. Heatmaps
### Using imshow() for visualization.

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.show()
```

### Understanding imshow() in Matplotlib
### The plt.imshow() function is used to display 2D data as an image, where each value in the array is represented by a color in the image. It's commonly used to visualize matrices, heatmaps, or images

```python
data = np.random.rand(10, 10)  # Generate a 10x10 matrix with random values between 0 and 1
plt.imshow(data, cmap='coolwarm', interpolation='nearest')  # Display the matrix as an image
plt.colorbar()  # Add a colorbar to indicate value ranges
plt.show()  # Display the plot
```

```python
data = np.random.rand(10, 10)  Creates a 10x10 matrix of random values between 0 and 1.
plt.imshow(data, cmap='coolwarm', interpolation='nearest'):
cmap='coolwarm'  Sets the colormap (colors representing values).
interpolation='nearest'  Sets how pixel values are interpolated (nearest neighbor for sharp pixel edges).
plt.colorbar()  Adds a colorbar beside the image to show the mapping between colors and values.
plt.show()  Displays the image and the plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.rand(10, 10)

# Display as image with 'hot' colormap and interpolation
plt.imshow(data, cmap='hot', interpolation='bilinear')  # Smooth interpolation

# Add a colorbar to show value ranges
plt.colorbar()

# Add title and labels
plt.title("Random Heatmap")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
####  Summary of Matplotlib
Matplotlib is a powerful Python plotting library used for creating static, animated, and interactive visualizations in 2D and 3D. It provides a range of features to create high-quality graphs, charts, and plots for data analysis and presentation.

Key Components:
pyplot: The most common module used to create basic plots (e.g., plt.plot(), plt.show(), plt.scatter()).
Axes: The region in a figure where data is plotted. Each axis (X and Y) represents one dimension of the data.
Figure: The entire window or image that contains the plot.
Basic Plot Types:
Line Plot: plt.plot(x, y)
Scatter Plot: plt.scatter(x, y)
Bar Plot: plt.bar(x, height)
Histogram: plt.hist(data)
Pie Chart: plt.pie(sizes)
Heatmap: plt.imshow(data)
Key Features:
Customization: You can customize titles, axis labels, colors, line styles, markers, and more.
Colormaps: Control how values are represented by colors (e.g., 'viridis', 'coolwarm').
Subplots: Display multiple plots in one figure (plt.subplot()).
3D Plots: Create 3D visualizations using Axes3D and ax.plot_surface().
Annotations: Add text, arrows, and annotations to highlight important points (ax.annotate()).
Saving Plots: Save figures in various formats (e.g., PNG, PDF) with plt.savefig().
Customization Options:
Color: Customize plot colors using colormaps (cmap) and specific colors for lines and markers.
Line Styles: Use solid, dashed, dotted, etc., for line styles.
Markers: Control the appearance of points in plots ('o', '^', 'x').
Legends: Add legends to explain different data series (plt.legend()).
Grids: Enable grids for better readability (plt.grid()).
Advanced Features:
Multiple Axes: Use twinx() and twiny() to add secondary axes.
Interactivity: Support for interactive plots via Jupyter Notebooks.
Animations: Create dynamic plots using FuncAnimation.
Common Use Cases:
Data Analysis: Visualize trends, distributions, and relationships.
Scientific Visualization: Present complex data like heatmaps, surface plots, and 3D models.
Presentation: Create high-quality graphs for reports and presentations.
```
